```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importando o conjunto de dados do Titanic
url_titanic = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = pd.read_csv(url_titanic)

# Exibindo as primeiras linhas do conjunto de dados
print(df_titanic.head())

# Informações básicas sobre o conjunto de dados
print(df_titanic.info())

# Estatísticas descritivas
print(df_titanic.describe())

# Contagem de sobreviventes e não sobreviventes
sns.countplot(x='Survived', data=df_titanic)
plt.title('Contagem de Sobreviventes e Não Sobreviventes')
plt.show()

# Distribuição das idades dos passageiros
sns.histplot(data=df_titanic, x='Age', bins=20, kde=True)
plt.title('Distribuição das Idades dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.show()

# Distribuição de passageiros por classe
sns.countplot(x='Pclass', data=df_titanic)
plt.title('Distribuição de Passageiros por Classe')
plt.xlabel('Classe')
plt.ylabel('Contagem')
plt.show()

# Distribuição de passageiros por sexo
sns.countplot(x='Sex', data=df_titanic)
plt.title('Distribuição de Passageiros por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()

# Distribuição de sobreviventes por classe
sns.countplot(x='Pclass', hue='Survived', data=df_titanic)
plt.title('Distribuição de Sobreviventes por Classe')
plt.xlabel('Classe')
plt.ylabel('Contagem')
plt.show()

# Distribuição de sobreviventes por sexo
sns.countplot(x='Sex', hue='Survived', data=df_titanic)
plt.title('Distribuição de Sobreviventes por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```
